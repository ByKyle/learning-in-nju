```python
import pandas as pd

titanic_file_path = '../TitanicMachineLearningfromDisaster'
data_file = open(titanic_file_path + '/train.csv')
```

```python
train_data = pd.read_csv(data_file)
# train_data.describe()
train_data.describe()
```

```python
train_data = train_data.dropna(axis=0)

features = ['PassengerId', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']
x = train_data[features]
x.describe()
```

```python
x.head()
```

```python
x.head(n=6)
```

```python
y = train_data['Survived']
```

```python
from sklearn.tree import DecisionTreeRegressor

model = DecisionTreeRegressor(random_state=1)

model.fit(x, y)
```

```python
model.predict(x.head())
```
