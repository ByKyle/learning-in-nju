# <center>读书笔记2 人件</center>
## 第一部分：人力资源
### 此时此刻，一个项目正在走向失败
    在绝大多数失败项目中，没有一个是因单纯的技术问题导致失败的。
   > &emsp;&emsp;我们工作中的问题更多属于社会学范畴，而非技术范畴。  
   > &emsp;&emsp;我们的成功源自于所有参与者良好的人与人之间的互动，我们的失败则归因于这种互动的缺失。  
    
### 干酪汉堡，做一个，卖一个
    —— 这种思维观念在开发领域是致命的。  
    
   - 错误在所难免，我们应当容许这种错误的存在。
   - 管理并不是向人们施压。即使向人们施压可以增加短期产出，长远来看是无效的。
   - 员工的独特个性使得项目团队产生了化学反应，是团队充满活力与高效的源泉。
   - 催化剂很重要，因为项目总是会处于不断变化的状态。一个能够让项目更稳定的人抵得上两个做事的人。
   
### 维也纳在等你
    压力不会让人工作得更好——只是工作的更快。
    为了工作快点，他们不得不牺牲产品质量以及自身的工作体验。
    
### 质量——如果时间允许
    ——这种策略会导致产品不会有任何质量可言  
        任何情况下都要追求高质量。
        我们通常倾向于将我们的自信与生产处的产品质量紧紧关联。采取任何可能牺牲产品质量的行动都可能挑起员工反对你的情绪。而管理人员设定的不可达到的期限威胁着产品的质量——当员工处于极度的时间压力之下时，质量就开始被牺牲了。
        质量，远远不只是最终用户的要求，而是达到高产能的一种方法。
    
### 苦杏素——并不能治愈任何疾病，却因宣传被癌症患者趋之若鹜
        不少管理人员很容易成为某种号称能够提高产能的技术苦味素的受害者。他们买到的技术其实缺乏任何客观证据的支撑。因为需求如此迫切，他们就会忽略对证据的审视。  
        管理者的作用不是让大家去工作，而是创造环境，让大家可以顺利开展工作。
    
## 第二部分：办公环境
### 家具警察
        只要员工还拥挤在嘈杂、低效、干扰不断的环境里，任何除了环境外的改进都是徒劳的。不能妄图规划一个完全整齐划一的工作环境，人们倾向于把自己的地方打造的带有自己的风格，以方便自己工作。
